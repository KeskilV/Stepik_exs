Необходимо посчитать средний возраст и заработную плату работников в зависимости от специализации. Отсортировать полученный результат по возрастанию среднего возраста.

```python
from ast import literal_eval as le
import pandas as pd
df = le(input())
res = pd.DataFrame(df)#.sort_values('salary',ascending = False).reset_index(drop = True)[['name', 'age', 'salary']].head(10)
#res.columns.name = res.index.name
#res.index.name = None
res.rename_axis('RATING', axis='columns', inplace=True)
print(res.groupby(['specialization'])['age', 'salary'].mean().sort_values(by = 'age'))
```
